```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import os
import random
```

```python
seed = 42
np.random.seed(seed)
random.seed(seed)
```

```python
train_features = np.load('data/train_features.npy')
train_labels = np.load('data/train_labels.npy')
test_features = np.load('data/test_features.npy')
```

```python
def plot_images(images, labels=None, rows=1, cols=10, figsize=(15, 15)):
    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    fig.subplots_adjust(hspace=0.2, wspace=0.1)  # Adjust horizontal and vertical spacing

    for i, ax in enumerate(axes.flat):
        if images.shape[-1] == 1:
            ax.imshow(images[i].reshape(32, 32), cmap='gray')
        else:
            ax.imshow(images[i])

        ax.axis('off')  # turn off axes

        if labels is not None:
            ax.set_title(f"Label: {labels[i]}", fontsize=16) 

    plt.show()
```

```python
plot_images(train_features, train_labels, rows=4, cols=4)
```

```python
plot_images(test_features, rows=4, cols=4) 
```
